# House Price Prediction Using Machine Learning

## Objective
The goal of this project is to predict house prices using machine learning techniques. Features such as area, bedrooms, and bathrooms are used to predict house prices.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Loading Dataset

The dataset is loaded using pandas.

```python
data = pd.read_csv("House price prediction dataset.csv")

data.head()
```

## Data Preprocessing

Important columns are selected for prediction.

```python
X= data[['Area', 'Bedrooms', 'Bathrooms', 'Floors', 'YearBuilt']]

y= data['Price']
```

## Model Training

Linear Regression is used to train the machine learning model.

```python
model = LinearRegression()

model.fit(X,y)
```

## Predictions

The model predicts house prices using the selected features.

```python
prediction = model.predict(X)

print(prediction[:5])
```

## Model Evaluation

MAE and RMSE are used to evaluate the model performance.

```python
mae = mean_absolute_error(y, prediction)

rmse = mean_squared_error(y, prediction) ** 0.5

print("MAE:", mae)

print("RMSE:", rmse)
```

## Data Visualization

The graph below compares actual house prices with predicted house prices.

```python
plt.figure(figsize=(10,5))

plt.plot(y.values, label='Actual Prices')

plt.plot(prediction, label='Predicted Prices')

plt.title("Actual vs Predicted House Prices")

plt.xlabel("Houses")

plt.ylabel("Price")

plt.legend()

plt.show()
```

## Final Insights

The Linear Regression model successfully predicted house prices using historical housing data. The predicted prices were close to the actual prices, showing that machine learning can identify patterns in housing data.
